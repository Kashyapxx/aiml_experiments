```python
import pandas as pd

# Function to import data from a CSV file
def import_data(file_path):
    try:
        data = pd.read_csv('data set.csv')
        return data
    except Exception as e:
        print(f"Error: {e}")
        return None

# Function to export data to a CSV file
def export_data(data, output_file_path):
    try:
        data.to_csv(output_file_path, index=False)
        print(f"Data exported successfully to {output_file_path}")
    except Exception as e:
        print(f"Error: {e}")

# Function to display dataset details
def display_details(data):
    if data is not None:
        print("Dataset Information:")
        print(f"Number of Rows: {data.shape[0]}")
        print(f"Number of Columns: {data.shape[1]}")
        print("\nFirst Five Rows:")
        print(data.head())
        print(f"\nSize of the Dataset (Total Elements): {data.size}")
        print("\nNumber of Missing Values in Each Column:")
        print(data.isnull().sum())

        print("\nSummary Statistics for Numerical Columns:")
        print(data.describe())

        numerical_cols = data.select_dtypes(include=['number'])
        if not numerical_cols.empty:
            print(f"\nSum of Numerical Columns:\n{numerical_cols.sum()}")
            print(f"\nMean of Numerical Columns:\n{numerical_cols.mean()}")
            print(f"\nMin Values of Numerical Columns:\n{numerical_cols.min()}")
            print(f"\nMax Values of Numerical Columns:\n{numerical_cols.max()}")
        else:
            print("\nNo numerical columns found.")
    else:
        print("No data to display.")

# Example usage
file_path = 'your_dataset.csv'  # Replace with your file path
output_file_path = 'exported_data.csv'

# Importing the data
data = import_data(file_path)

# Displaying dataset details
display_details(data)

# Exporting the data to a new CSV file
export_data(data, output_file_path)
```
